```python
import os
import pandas as pd
import sqlite3
import shutil

current_dir = os.getcwd()
db_path = os.path.join(current_dir, 'worldcities.db')
connection = sqlite3.connect(db_path)
print('Connected to', db_path)
```

```python
def run_query(q):
    print(pd.read_sql_query(q, connection))

# Örnek: tablodaki ilk 5 satırı göster
run_query('SELECT city, country, population FROM cities LIMIT 5')
```

```python

# UPDATE örneği
connection.execute("UPDATE cities SET population = ? WHERE city = ?", (20000, 'YeniSehir'))
connection.commit()
run_query("SELECT city, population FROM cities WHERE city='YeniSehir'")

# DELETE örneği
connection.execute("DELETE FROM cities WHERE city = ?", ('YeniSehir',))
connection.commit()
run_query("SELECT city FROM cities WHERE city='YeniSehir'")
```

```python
# Transaction örneği
try:
    connection.execute('BEGIN')
    connection.execute("INSERT INTO cities (city, country, population) VALUES (?, ?, ?)", ('TxCity1','Testland',100))
    connection.execute("INSERT INTO cities (city, country, population) VALUES (?, ?, ?)", ('TxCity2','Testland',200))
    connection.commit()
except Exception as e:
    connection.rollback()
    print('Rolled back:', e)
finally:
    run_query("SELECT city, population FROM cities WHERE country='Testland'")
```

```python
# Yedekleme (dosya kopyası)
bak_path = db_path + '.bak'
shutil.copy(db_path, bak_path)
print('Backup created:', bak_path)
```

```python
# Bağlantıyı kapat
connection.close()
print('Connection closed.')
```
